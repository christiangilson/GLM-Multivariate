## Looking at the dose-response example from L1.

> That example showed it to be a binomial distribution question;

> And it showed it to be a logistic regression question

> Thus the **logit link** function was used\*

Let's start by loading in some data:

\* NB: the default link function used by R is the canonical link function

```r
dose <- c(1.6907, 1.7242, 1.7552, 1.7842, 1.8113, 1.8369, 1.861, 1.8839)
number <- c(59, 60, 62, 56, 63, 59, 62, 60)
killed <- c(6, 13, 18, 28, 52, 53, 61, 60)
```

### Recall that we changed our random variable from being the number killed to being the percentage of the total killed

```r
pc_killed <- killed / number
```

```r
# centering the dose values
dose <- dose - mean(dose)
```

## Linear prediction, $\eta_{i} = \beta_{1} + \beta_{2} x_{i}$

> $\beta_{1}$ represents the log odds of success at x=0, since $g(\mu_{i}) = log(\dfrac{\mu_{i}}{1 - \mu_{i}}) = \eta_{i} = \beta_{1} + \beta_{2} x_{i}$

> By subtracting the mean dose, the intercept now represents the mean fraction of insects killed provided the mean dose, rather than a zero dose

```r
df.pesticide <- data.frame(dose, number, killed, pc_killed)

df.pesticide
```

## Fitting the model

> Note the weights here are for the $a_{i}(\phi)$ part of the generalised exponential family equation

> $a_{i}(\phi) = \phi / w_{i}$, where $w_{i}$ are the weights for observation $i$, and $\phi$ is a constant representing dispersion.

> i.e. the $a_{i}(\phi)$ term allows the variance of the r.v. to increase based on, say, an increasing mean, or any other factor

> and hense why $\text{Var}(Y_{i}) = a_{i}(\phi) b''(\theta)$

> the number of insects here are our weights - this fell out of our equations in the lecture, where $\phi = 1$

```r
#help(family)
```

```r
# the canonical link function for the binomial distribution is the logit link,
# so this would be used automatically, but we're just entering it here to be explicit
glm.pesticide.logit <- glm(pc_killed ~ dose, family = binomial(link='logit')
                           ,weights = number, data = df.pesticide)
```

```r
summary(glm.pesticide.logit)
```

```r
anova(glm.pesticide.logit)
```

> Has a scaled deviance of 11.23 on 6 degrees of freedom

> This has a p-value of 0.082

```r
pchisq(11.23, 6, lower.tail = FALSE)
```

## Plotting the logistic regression model

```r
x <- seq(min(dose)-0.1, max(dose)+0.1, 0.005)
pred.logit <- predict(glm.pesticide.logit, data.frame(dose=x), type="response")

plot(dose, pc_killed)
lines(x, pred.logit)
```

## Plotting the residuals

```r
dev.residuals <- resid(glm.pesticide.logit, type='deviance')
arcsin.root.fitted <- asin(sqrt(fitted(glm.pesticide.logit)))

plot(dev.residuals, arcsin.root.fitted)
```

## Ch. 6: Fitting different link functions: probit and complimentary log-log

## **Probit**

```r
glm.pesticide.probit <- glm(pc_killed ~ dose, family = binomial(link='probit')
                            ,weights = number, data=df.pesticide)
```

```r
summary(glm.pesticide.probit)
```

```r
anova(glm.pesticide.probit)
```

```r
pchisq(10.12, 6, lower.tail=F)
```

## Probit Plots

```r
probit.fitted <- fitted(glm.pesticide.probit)
probit.dev.resid <- resid(glm.pesticide.probit, type='deviance')

plot(probit.fitted, probit.dev.resid)
```

## Trend in the mean - Not adequete.

```r
x <- seq(-0.1, 0.1, 0.001)
pred.probit <- predict(glm.pesticide.probit, data.frame(dose=x), type='response')

plot(df.pesticide$dose, df.pesticide$pc_killed)
lines(x, pred.probit)
```

## **Complimentary Log-Log**

```r
glm.pesticide.cloglog <- glm(pc_killed ~ dose, family=binomial(link='cloglog')
                             ,weights=number, data=df.pesticide)
```

```r
summary(glm.pesticide.cloglog)
```

```r
anova(glm.pesticide.cloglog)
```

```r
pchisq(3.4464, 6, lower.tail=F)
```

## Super D. Can't reject the null. Happy with params and the link. Good fitting model.

```r
cloglog.fitted <- fitted(glm.pesticide.cloglog)
cloglog.dev.res <- resid(glm.pesticide.cloglog, type='deviance')

plot(cloglog.fitted, cloglog.dev.res)
```

```r
pred.cloglog <- predict(glm.pesticide.cloglog, data.frame(dose=x), type='response')

plot(df.pesticide$dose, df.pesticide$pc_killed)
lines(x, pred.cloglog)
```

## Good lookin' model. Nice residuals. Good deviance. And don't forget about the AIC: the AIC for the cloglog link was the best of the bunch. Endorse this model.
