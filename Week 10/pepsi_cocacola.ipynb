```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/christiangilson/Desktop/Hivemind/Quandl/Quandl_Equity_Prices.csv')
```

```python
df_pepsi = df.loc[df['ticker'] == 'PEP']
```

```python
df_coke = df.loc[df['ticker'] == 'KO']
```

```python
df_coke['return'] = ((df_coke['split_ratio'] * (df_coke['close'] + df_coke['ex-dividend'])) / df_coke['close'].shift(1)) - 1.0
df_pepsi['return'] = ((df_pepsi['split_ratio'] * (df_pepsi['close'] + df_pepsi['ex-dividend'])) / df_pepsi['close'].shift(1)) - 1.0
```

```python
df_coke
```

```python
df_pair = df_coke.merge(df_pepsi, on='date', how='inner', suffixes=('_coke','_pepsi'))
```

```python
df_pair = df_pair[['return_coke','return_pepsi']]
```

```python
df_pair = df_pair.loc[pd.isna(df_pair['return_coke']) == False]
df_pair = df_pair.loc[pd.isna(df_pair['return_pepsi']) == False]
```

```python
df_pair.to_csv('coke_pepsi_paired_returns.csv', index=None)
```
